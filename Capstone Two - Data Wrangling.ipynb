```python
# Create histograms
fig, axes = plt.subplots(nrows=6, ncols=4, figsize=(18, 18))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    dairy_data[col].hist(ax=axes[i], bins=30, color='skyblue', edgecolor='black')
    axes[i].set_title(col, fontsize=10)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    # Rotate x labels if numeric range is wide
    axes[i].tick_params(axis='x', labelrotation=30)

# Remove unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(pad=2.0)
plt.show()
```

```python
# Function to calculate IQR and visualize
def visualize_and_detect_outliers(df, columns):
    fig, axes = plt.subplots(nrows=len(columns)//3 + 1, ncols=3, figsize=(18, 4*len(columns)//3))
    axes = axes.flatten()

    for i, col in enumerate(columns):
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        # Plot boxplot
        sns.boxplot(x=df[col], ax=axes[i], color='skyblue')
        axes[i].axvline(lower, color='red', linestyle='--', label='Lower Bound')
        axes[i].axvline(upper, color='orange', linestyle='--', label='Upper Bound')
        axes[i].set_title(f"{col}")
        axes[i].legend()

    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()

# Run visualization
visualize_and_detect_outliers(dairy_data, num_cols)
```

```python
# Use IQR (Interquartile Range) to find extreme revenue values
Q1 = dairy_data['Approx. Total Revenue(INR)'].quantile(0.25)
Q3 = dairy_data['Approx. Total Revenue(INR)'].quantile(0.75)
IQR = Q3 - Q1

dairy_data[
    (dairy_data['Approx. Total Revenue(INR)'] < Q1 - 3 * IQR) |
    (dairy_data['Approx. Total Revenue(INR)'] > Q3 + 3 * IQR)
]
```

```python
# Compute revenue per unit sold
dairy_data['Revenue_per_Unit'] = dairy_data['Approx. Total Revenue(INR)'] / dairy_data['Quantity Sold (liters/kg)']

# See if outliers have abnormally high revenue per unit
dairy_data[['Brand', 'Product Name', 'Revenue_per_Unit']].sort_values('Revenue_per_Unit', ascending=False).head(10)
```

```python
# Revenue per unit should be within a reasonable range
# Very large or very small values may indicate a unit error (e.g., liters vs kilograms)

dairy_data['Revenue_per_Unit'] = (
    dairy_data['Approx. Total Revenue(INR)'] / dairy_data['Quantity Sold (liters/kg)']
)

dairy_data[
    dairy_data['Revenue_per_Unit'] > dairy_data['Revenue_per_Unit'].mean() * 100
]
```

###  Outlier Handling with IQR method Summary

- **Step 1 – Outlier Detection:**  
  - Used the **Interquartile Range (IQR)** method to identify extreme revenue values.  
  - Records with `Approx. Total Revenue (INR)` beyond **3 × IQR** from the upper quartile were flagged as potential outliers.

- **Step 2 – Validation using Revenue per Unit Sold:**  
  - Created a new column:  
    {Revenue per Unit} = {Approx. Total Revenue (INR)}/{Quantity Sold (liters/kg)} 
  - Checked for unusually high unit revenues to confirm whether the total revenue outliers were genuine.

- **Step 3 – Result:**  
  - Top brands (e.g., *Sudha*, *Amul*, *Warana*) showed consistent **Revenue per Unit** around ₹100–₹105.  
  - This consistency confirmed that high total revenues resulted from **large sales volumes**, not data errors.

- **Step 4 – Action Taken:**  
  - No records were removed.  
  - All outliers were **retained for further analysis** as valid high-performing entries.

```python
# Define IQR for Shelf Life
Q1 = dairy_data['Shelf Life (days)'].quantile(0.25)
Q3 = dairy_data['Shelf Life (days)'].quantile(0.75)
IQR = Q3 - Q1

# Find outliers
shelf_outliers = dairy_data[
    (dairy_data['Shelf Life (days)'] < Q1 - 1.5 * IQR) |
    (dairy_data['Shelf Life (days)'] > Q3 + 1.5 * IQR)
]

# Inspect the product types of those outliers
shelf_outliers.groupby(['Product Name', 'Storage Condition']).size().reset_index(name='Count')
```

##### So although the model flags them as outliers statistically, they are not errors — they just represent long-lasting products that behave differently from others like milk or curd.

```python
# Check impossible values: values that cannot exist in real life
# Example: negative shelf life, zero price, or zero production

dairy_data[
    (dairy_data['Price per Unit'] <= 0) |
    (dairy_data['Shelf Life (days)'] <= 0) |
    (dairy_data['Quantity (liters/kg)'] <= 0)
]
```

```python
# Sales should not exceed stock or production
# Logical check: Quantity Sold cannot be greater than Quantity Produced

dairy_data[dairy_data['Quantity Sold (liters/kg)'] > dairy_data['Quantity (liters/kg)']]
```

#### Verify numeric relationships

```python
# Verify Revenue = Quantity Sold × Price per Unit (sold)
dairy_data['Calculated_Revenue'] = (
    dairy_data['Quantity Sold (liters/kg)'] * dairy_data['Price per Unit (sold)']
).round(2)

# Compare actual vs calculated revenue
revenue_check = dairy_data[
    ~np.isclose(dairy_data['Approx. Total Revenue(INR)'], dairy_data['Calculated_Revenue'], rtol=1e-3)
]

print(f"Number of mismatches in revenue calculation: {len(revenue_check)}")
```

```python
dairy_data[['Approx. Total Revenue(INR)', 'Calculated_Revenue']].head(10)
```

```python
# Verify Total Value = Quantity × Price per Unit
dairy_data['Calculated_Total_Value'] = (
    dairy_data['Quantity (liters/kg)'] * dairy_data['Price per Unit']
).round(2)

# Compare actual vs calculated total value
total_value_check = dairy_data[
    ~np.isclose(dairy_data['Total Value'], dairy_data['Calculated_Total_Value'], rtol=1e-3)
]

print(f"Number of mismatches in total value calculation: {len(total_value_check)}")
```

```python
dairy_data[['Total Value', 'Calculated_Total_Value']].head(10)
```

### Verification Summary

- **Revenue Validation**
  - Formula checked: `Quantity Sold × Price per Unit (sold) = Approx. Total Revenue (INR)`
  - All records matched — no discrepancies found.

- **Total Value Validation**
  - Formula checked: `Quantity × Price per Unit = Total Value`
  - All records matched — calculations consistent.

Therefore, both `Revenue` and `Total Value` columns are **mathematically correct**.
